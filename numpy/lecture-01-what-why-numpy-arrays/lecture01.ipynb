## Code Example 1 : Create arrays and inspect them

```python
import numpy as np

a = np.array([10, 20, 30, 40, 50])

print("Array:", a)
print("dtype:", a.dtype)
print("shape:", a.shape)
print("ndim:", a.ndim)
print("size:", a.size)
```

## Code Example 2 : dtype differences

```python
a = np.array([1, 2, 3])
print("Default int dtype:", a.dtype)

b = np.array([1.0, 2.0, 3.0])
print("Default float dtype:", b.dtype)

c = np.array([1, 2, 3], dtype=np.float32)
print("Forced float32:", c.dtype)

d = np.array([0, 1000, 65535], dtype=np.uint16)
print("uint16 array:", d)
print("Max possible uint16 value:", np.iinfo(np.uint16).max)
```

## Code Example 3 : 2D arrays (matrices)

```python
image = np.array([
    [10, 20, 30],
    [40, 50, 60],
    [70, 80, 90]
])

print("2D array:\n", image)
print("shape:", image.shape)
print("ndim:", image.ndim)
print("size:", image.size)

rows = image.shape[0]
cols = image.shape[1]
print(f"Rows: {rows}, Columns: {cols}")
```

## Code Example 4 : Speed comparison: Python list vs NumPy array

```python
import time

size = 1_000_000

py_list = list(range(size))
start = time.time()
result = [x * 2 for x in py_list]
end = time.time()
print(f"Python list: {(end - start)*1000:.2f} ms")

np_array = np.arange(size)
start = time.time()
result = np_array * 2
end = time.time()
print(f"NumPy array: {(end - start)*1000:.2f} ms")
```

## Code Example 5 : Useful array creation functions

```python
zeros = np.zeros((3, 4))
print("Zeros:\n", zeros)

ones = np.ones((2, 3))
print("Ones:\n", ones)

x = np.arange(0, 10, 2)
print("Arange:", x)

energies = np.linspace(1, 6, 5)
print("Energies:", energies)

template = np.array([[1, 2], [3, 4]])
background = np.zeros_like(template)
print("zeros_like:\n", background)
```
